```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np 
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_excel('LedaHealthInternship.xlsx')
a=df.loc[df['Are you a survivor of sexual assault?']=='No'].index
for each in a:
    df.drop(each,inplace=True)
```

## Covid_19 effects on Sexuall Aggressiveness

```python
sns.countplot(df["Has your partner gotten more sexually aggressive since COVID-19's outbreak?"]);
```

## The plot comprehends that those having a partner didn't report there assualt to the Police.

```python
pd.crosstab(df['Do you have a partner? '],df['Did you report your assault to the police? ']).plot(kind='bar')
plt.title("Assault Reporting To police for Partner")
plt.xlabel("Partner")
plt.ylabel("Frequency of Report to Police")
plt.show()
```

## The plot below revelas that those who have experienced sexual assualt are still experiencing it.

```python
pd.crosstab(df['Are you currently experiencing sexual assault?'],df['Have you previously experienced sexual assault?']).plot(kind='bar')
plt.title("Sexual Assualt Past And Present ")
plt.xlabel("Are you currently experiencing sexual assault?")
plt.ylabel("Experienced")
plt.show() 
```

## It appears that COVID_19 has no effects on partner on partner abuse.

```python
pd.crosstab(df["Has your partner gotten more sexually aggressive since COVID-19's outbreak?"],df['Are you currently experiencing partner on partner abuse?']).plot(kind='bar')
plt.title("Covid_19_Effects On sexual assualt")
plt.xlabel("Are you currently experiencing sexual assault?")
plt.ylabel("Covid_19_ Effects on partner ")
plt.show() 
```

## The plot below shows that more  men are experiencing partner on partner abuse as compare to women while most of the people are not experiencing partner on partner abuse. 

```python
plt.figure(figsize=(15,20))
pd.crosstab(df["What gender do you identify as?"],df['Are you currently experiencing partner on partner abuse?']).plot(kind='bar')
plt.title("GenderWise Sexual assualt")
plt.xlabel("Gender")
plt.ylabel(" Abuse ")
plt.show() 
```

```python
cat_vars=['Ethnicity ']
for var in cat_vars:
    cat_list=pd.get_dummies(df[var],prefix=var)
    df=df.join(cat_list)
```

## The Plot below shows that a large portion white Ethnicity people do report there assualt to the police.

```python
sns.catplot(x="Did you report your assault to the police? ", y="Ethnicity _White",kind="violin", data=df);
```

# The following plot shows that those who have performed SEAK doesn't trust the police as much as compare to those who haven't performed SEAK.  

```python
pd.crosstab(df["Did you have a Sexual Assault Evidence Kit (SAEK) performed?"],df['Do you trust the police?']).plot(kind='bar')
plt.title("(SAEK) performed")
plt.xlabel("Police Trust")
plt.ylabel("SAEK")
plt.show() 
```
